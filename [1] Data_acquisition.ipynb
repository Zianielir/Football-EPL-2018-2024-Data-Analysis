```python
df = pd.concat([pd.read_csv(f'E0 ({i}).csv', parse_dates=['Date'], date_format='mixed', dayfirst=True) for i in range(1, 7)]).reset_index(drop=True)
df
```

```python
df.columns.values
```

```python
cols = ['Date', 'Time', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG',
       'FTR', 'HTHG', 'HTAG', 'HTR', 'Referee', 'HS', 'AS', 'HST', 'AST',
       'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR', 'B365H', 'B365D',
       'B365A', 'BWH', 'BWD', 'BWA', 'IWH', 'IWD', 'IWA', 'PSH', 'PSD',
       'PSA', 'WHH', 'WHD', 'WHA', 'VCH', 'VCD', 'VCA']# 'LBH',
 #      'LBD', 'LBA', 'SJH', 'SJD', 'SJA']
df = df.loc[:, cols]
df
```

```python
df.isna().sum()
```

```python
# Пропуски в Time: заполним медианным временем
# Остальные пропуски в коэффициентах букмекеров - их и не будет в финальном датафрейме и и они не учтутся в .mean
```

```python
df['Time'].value_counts()
```

```python
df.head()
```

```python
df['Time'].fillna('15:00', inplace=True)
```

```python
# найдем все коэффициенты на победу H, A или ничьей D

df_bet_H = df[[col for col in cols if col[-1] == 'H']]
df_bet_D = df[[col for col in cols if col[-1] == 'D']]
df_bet_A = df[[col for col in cols if col[-1] == 'A']]
```

```python
df['AvgH'] = df_bet_H.mean(axis=1)
df['AvgD'] = df_bet_D.mean(axis=1)
df['AvgA'] = df_bet_A.mean(axis=1)
```
