- Изначальные main данные
- Данные, которые спарсили по погоде

```python
import numpy as np
import pandas as pd

from datetime import datetime
from collections import Counter
```

## Main dataframe

```python
df = pd.read_csv('main_df.csv', parse_dates=['Date']).iloc[:, 1:]
df.head()
```

## Weather data

Данные собирались в 3 прохода по несколько запусков, поэтому соединим сначала каждый проход, а затем уже все в один дата фрейм

Первая тысяча:

```python
df_weather_1 = pd.concat([pd.read_csv(f'[3]parsed_weather:1000({i}).csv', parse_dates=['Date']) for i in range(2)]).iloc[:, 1:]
df_weather_1
```

```python
df_weather_2 = pd.concat([pd.read_csv(f'[1]parsed_weather1000:1500({i}).csv', parse_dates=['Date']) for i in range(5)]).iloc[:, 1:]
df_weather_2
```

```python
df_weather_3 = pd.concat([pd.read_csv(f'[2]parsed_weather1500({i}).csv', parse_dates=['Date']) for i in range(6)]).iloc[:, 1:]
df_weather_3
```

```python
df_weather = pd.concat([df_weather_1, df_weather_2, df_weather_3])
df_weather
```

__Убираем пропуски (их там много из-за особенностей сбора данных)__

```python
df_weather.isna().sum()
```

```python
df_weather.dropna(subset='Time', inplace=True)
```

```python
df_weather
```

```python
df_weather[df_weather['Condition'].isna() == True]
```

```python
df_weather[(df_weather.Date == '2023-05-20') & (df_weather.Location == 'EGLC')]
```

Здесь вообще проблема с данными по погоде. Почему-то резко стало 0 градусов. Скорее всего это ошибка, поэтому было решено дропнуть данные об этом дне в данной локации

```python
df_weather = df_weather[~((df_weather.Date == '2023-05-20') & (df_weather.Location == 'EGLC'))]
df_weather
```

__Сортируем по датам__

```python
df_weather.sort_values(by='Date', ascending=False).reset_index(drop=True)
```

```python
df_weather.Time.value_counts()
```

```python
df_weather.loc[:, 'Time'] = df_weather.loc[:, 'Time'].apply(lambda x: x if x is None else datetime.strptime(x, '%I:%M %p').strftime('%H:%M'))
df_weather
```

__Температура__

```python
df_weather.loc[:, 'Temperature'] = df_weather.loc[:, 'Temperature'].apply(lambda x: int(x[:-2]))
df_weather
```

__Влажность__

```python
df_weather.loc[:, 'Humidity'] = df_weather.loc[:, 'Humidity'].apply(lambda x: int(x[:-2]))
df_weather
```

__Скорость ветра__

```python
df_weather.loc[:, 'Wind Speed'] = df_weather.loc[:, 'Wind Speed'].apply(lambda x: int(x[:-4]))
df_weather
```

### Elo ratings

```python
df_elo = pd.read_csv('elo_ratings_all_teams', parse_dates=['Date']).iloc[:, 1:]
df_elo
```

## Соединяем

```python
df_main_elo = df.merge(df_elo, on=['Date', 'Time', 'Location', 'HomeTeam', 'AwayTeam'], how='inner')
df_main_elo
```

```python
df_main = df_main_elo.merge(df_weather, on=['Date', 'Location'], how='left')
df_main
```

```python
df_main[df_main['Time_y'].isna()]
```

```python
df_main.isna().sum()
```

```python
time_x = pd.to_datetime(df_main['Time_x'], format='mixed')
time_y = pd.to_datetime(df_main['Time_y'], format='mixed')
```

```python
mask_time = (time_x - time_y >= '-1 days +22:00:00') & (time_x - time_y <= '0 days 00:30:00')
mask_nans = df_main['Time_y'].isna()

df_main = df_main[mask_time | mask_nans]
df_main
```

__Усредняем показатели погоды, чтобы было одна строка = один матч__

```python
df_main_grouped = df_main.groupby(['Date', 'HomeTeam'])[['Temperature', 'Humidity', 'Wind Speed', 'Condition']].agg({'Temperature': lambda x: np.nanmean(x),
                                                                                                'Humidity': lambda x: np.nanmean(x),
                                                                                                'Wind Speed': lambda x: np.nanmean(x),
                                                                                                'Condition': lambda x: Counter(x).most_common()[0][0]})
df_main_grouped.sort_index(ascending=False)
```

```python
df_final = df_main_elo.merge(df_main_grouped, on=['Date', 'HomeTeam'], how='left')
df_final
```

### Разбираемся с оставшимися nan

```python
df_final.isna().sum() 
```

Их не так много, поэтому просто заполним количественные признаки медианой (потому что могли быть выбросы), а качественный - модой, потому что скорее всего она бы там и была

```python
from sklearn.impute import SimpleImputer

# для вещественных признаков: численная медиана
mis_replacer = SimpleImputer(strategy='median')

df_final_no_nan = df_final.iloc[:, :-4]

df_final_real_nan = pd.DataFrame(
    data=mis_replacer.fit_transform(df_final.iloc[:, -4:-1]), columns=df_final.iloc[:, -4:-1].columns)

# для категориального мода 'Fair'
df_final_cond = df_final['Condition'].fillna('Fair')

# объединяем
df_final = pd.concat([df_final_no_nan, df_final_real_nan, df_final_cond], axis=1)
```

## Выгружаем файл

```python
df_final.to_csv('EPL_18_24_data.csv')
```
